```python
import numpy as np
```

```python
def outliers(df, i):
    q1, q2, q3 = np.quantile(df[i], [0.25, 0.5, 0.75])
    iqr = q3 - q1
    upper_whisker = q3 + (1.5 * iqr)
    lower_whisker = q1 - (1.5 * iqr)
    percentage = (len(df[(df[i] > upper_whisker) | (df[i] < lower_whisker)])/len(df)) * 100
    print(f'Outliers percentage of {i} : {percentage}')
```

```python
def exploratory_data_analysis(df1, target_variable):
    df = df1.drop(target_variable, axis = 1)
    range = df.shape
    columns = df.columns
    null_values = sum(df.isnull().sum())
    duplicated_values = df.duplicated().sum()
    numerical_columns = []
    categorical_columns = []
    threshold = 20
    for i in columns:
        l = len(df[i].unique())
        if l > threshold:
            numerical_columns.append(i)
        else:
            categorical_columns.append(i)
            
    unique_values_of_target_variable = df1[target_variable].unique()
    percentage = []
    for i in unique_values_of_target_variable:
        percentage.append((len(df1[df1[target_variable] == i])/len(df1)) * 100)
    
    print(f'Imbalanced dataset : {percentage}')
    
    print(f"""Range : {range}\n
    columns : {columns}\n
    Null_Values : {null_values}\n
    duplicated_values_count : {duplicated_values}\n 
    numerical_columns : {numerical_columns}\n
    categorical_columns : {categorical_columns}""")
    print('\n')
    
    for i in numerical_columns:
       outliers(df=df, i=i)
    
```

```python
exploratory_data_analysis(train_df, 'default payment next month')
```

```python
exploratory_data_analysis(test_df)
```
